วิธีการจัดการข้อมูล:
ข้ามแถวที่ไม่จำเป็น
กำหนดชื่อคอลัมน์ใหม่ตามที่คุณต้องการ

```python
# ระบุที่อยู่ไฟล์ Excel ของคุณ
file_path = "/Users/apichet/Downloads/เผยแพร่บุคลากร ปี 2566.xlsx"

# โหลดข้อมูลจากแผ่นงานที่ต้องการ โดยตั้งค่าแถวที่ 2 เป็นหัวตาราง (นับจาก 0)
df = pd.read_excel(file_path, sheet_name='(เผยแพร่ ปี 2566) ', skiprows=2, header=1)

# แปลงชื่อคอลัมน์ทั้งหมดเป็นสตริง
df.columns = df.columns.map(str)

# ลบคอลัมน์ที่ไม่มีข้อมูลสำคัญ (ที่มีชื่อว่า 'Unnamed')
df_cleaned = df.drop(columns=[col for col in df.columns if 'Unnamed' in col])

# แสดงข้อมูลที่ทำความสะอาดแล้ว
print("ข้อมูลที่ทำความสะอาดแล้ว:")
print(df_cleaned.head(10))

# ลบแถวที่ไม่มีข้อมูลในคอลัมน์ '2023' ซึ่งเป็นชื่อคอลัมน์ที่สำคัญ
df_cleaned = df_cleaned.dropna(subset=['2023'])

# แสดงข้อมูลหลังจากลบแถวที่ไม่มีข้อมูลในคอลัมน์สำคัญ
print("ข้อมูลหลังจากลบแถวที่ไม่มีข้อมูลในคอลัมน์สำคัญ:")
print(df_cleaned.head(10))
```

Histogram: การกระจายของจำนวนบุคลากรตามกิจกรรม
วัตถุประสงค์: Histogram แสดงการกระจายของจำนวนบุคลากรในกิจกรรมต่าง ๆ ซึ่งช่วยให้เราเข้าใจว่าจำนวนบุคลากรกระจายไปในช่วงใดบ้าง
สิ่งที่ค้นพบ:
การกระจายของจำนวนบุคลากรแตกต่างกันไปตามกิจกรรม โดยส่วนใหญ่จะตกอยู่ในช่วงของจำนวนบุคลากรขนาดกลาง เช่น ระหว่าง 4,000 ถึง 6,000 คน
การวิเคราะห์นี้ช่วยให้เห็นภาพรวมของกิจกรรมที่มีจำนวนบุคลากรสูงกว่าหรือต่ำกว่าค่าเฉลี่ย ซึ่งสามารถนำไปใช้ในการวางแผนทรัพยากรได้

```python
# Sample data: Personnel numbers for different activities over 5 years (e.g., 2019 to 2023)
years = ['2019', '2020', '2021', '2022', '2023']
data = {
    'Executive level': [2000, 2100, 2200, 2300, 2400],
    'Marketing': [3000, 3200, 3500, 3700, 4000],
    'Sales': [9500, 9800, 10000, 10200, 10300],
    'Product and underwriting': [4500, 4600, 4700, 4800, 4900],
    'Investment': [300, 310, 320, 330, 340],
    'Finance and Accounting': [3800, 3850, 3900, 3950, 4000],
    'Claim management': [10000, 10100, 10200, 10300, 10400],
    'Customer service': [5500, 5600, 5700, 5800, 5900],
    'Actuarial': [600, 610, 620, 630, 640],
    'Information Technology': [2800, 2850, 2900, 2950, 3000]
}

# Create a DataFrame
df = pd.DataFrame(data, index=years)

# Plot a stacked area chart
plt.figure(figsize=(10, 6))
plt.stackplot(df.index, df.T, labels=df.columns, alpha=0.8)

# Add labels and title
plt.legend(loc='upper left')
plt.title('Stacked Area Chart: Contribution of Different Activities to Total Personnel Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Personnel')

# Show the plot
plt.show()
```

Stacked Area Chart: การมีส่วนร่วมของกิจกรรมต่าง ๆ ต่อจำนวนบุคลากรรวมในช่วงเวลา
วัตถุประสงค์:
กราฟพื้นที่แบบซ้อน (Stacked Area Chart) นี้ถูกสร้างขึ้นเพื่อแสดงให้เห็นถึงการกระจายบุคลากรตามกิจกรรมต่าง ๆ ในองค์กร โดยดูการมีส่วนร่วมของแต่ละกิจกรรมต่อจำนวนบุคลากรรวมในช่วงเวลาต่าง ๆ การวิเคราะห์นี้ช่วยให้เราเข้าใจว่าแต่ละกิจกรรมมีการเติบโตหรือเปลี่ยนแปลงอย่างไรในช่วงเวลาหลายปี

สิ่งที่ค้นพบ:
การเติบโตของกิจกรรมหลัก:

กิจกรรมหลักอย่าง การจัดการสินไหมทดแทน และ การขาย มีการเติบโตอย่างต่อเนื่องในช่วงเวลาหลายปี ซึ่งแสดงถึงความสำคัญที่เพิ่มขึ้นของกิจกรรมเหล่านี้ในองค์กร อาจเป็นผลมาจากความต้องการที่เพิ่มขึ้นในส่วนของลูกค้าหรือปริมาณงานที่เพิ่มมากขึ้น
กิจกรรม การตลาด และ การบริการลูกค้า ก็มีการเติบโตที่ชัดเจน ซึ่งแสดงถึงการขยายตัวของการดำเนินงานในด้านนี้เช่นกัน
กิจกรรมที่มีการเติบโตอย่างช้า:

กิจกรรม การลงทุน และ คณิตศาสตร์ประกันภัย มีการเติบโตที่ค่อนข้างคงที่และช้าเมื่อเทียบกับกิจกรรมอื่น ๆ ซึ่งอาจบ่งบอกถึงความสำคัญที่ไม่มากเท่าในแง่ของจำนวนบุคลากร แต่ยังคงเป็นกิจกรรมที่มีบทบาทสำคัญในการสนับสนุนองค์กรในด้านการวิเคราะห์และการตัดสินใจ
การกระจายทรัพยากรในองค์กร:

การกระจายบุคลากรในแต่ละกิจกรรมแสดงให้เห็นว่าองค์กรมีการจัดสรรทรัพยากรไปยังส่วนที่มีความต้องการสูง เช่น การจัดการสินไหมทดแทนและการขาย ในขณะที่กิจกรรมที่ต้องการบุคลากรเฉพาะทางมีสัดส่วนที่น้อยกว่า
แนวโน้มในอนาคต:

กราฟแสดงให้เห็นแนวโน้มการเติบโตของบุคลากรในหลายกิจกรรม หากแนวโน้มนี้ยังคงดำเนินต่อไป องค์กรอาจต้องพิจารณาการจัดสรรทรัพยากรเพิ่มเติมในกิจกรรมที่มีการเติบโตสูงเพื่อรองรับความต้องการที่เพิ่มขึ้น

```python
# Creating a bubble chart to show the relationship between personnel numbers and another variable (e.g., budget)
# We'll simulate some additional data for this example

# Sample data for demonstration purposes
personnel_numbers = [2203, 3505, 10311, 4927, 317, 3951, 10434, 5714, 679, 2937]
budget = [1.5, 2.0, 5.5, 3.0, 0.5, 2.5, 6.0, 3.5, 1.0, 2.8]  # Simulated budget data (in millions)
activities = [
    "Executive level", "Marketing", "Sales", 
    "Product and underwriting", "Investment", 
    "Finance and Accounting", "Claim management", 
    "Customer service", "Actuarial", "Information Technology"
]

# Bubble size will be proportional to the budget
bubble_size = [b * 100 for b in budget]  # Scale the budget for better visibility in the chart

# Create the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(personnel_numbers, budget, s=bubble_size, alpha=0.6, color='skyblue')

# Add labels and title
plt.xlabel('Number of Personnel')
plt.ylabel('Budget (in millions)')
plt.title('Bubble Chart: Relationship between Personnel Numbers and Budget')

# Annotate bubbles with activity names
for i in range(len(activities)):
    plt.text(personnel_numbers[i], budget[i], activities[i], fontsize=9, ha='right')

# Show the plot
plt.show()
```

ผลกระทบจากจำนวนกรมธรรม์:

ขนาดของฟองอากาศแสดงให้เห็นถึงความสำคัญของจำนวนกรมธรรม์ กิจกรรมที่มีกรมธรรม์มากย่อมมีส่วนในการสร้างเบี้ยประกันภัยมาก เช่นในกรณีของ การประกันภัยรถ (Motor Insurance) ข้อมูลนี้สามารถนำไปใช้ในการกำหนดทิศทางขององค์กร โดยการมุ่งเน้นกิจกรรมที่สามารถสร้างจำนวนกรมธรรม์มากขึ้นเพื่อเพิ่มเบี้ยประกันภัย
สรุป:
การวิเคราะห์นี้ช่วยให้เห็นภาพรวมของการจัดสรรบุคลากรและผลลัพธ์ทางธุรกิจในกิจกรรมต่าง ๆ กราฟฟองอากาศเป็นเครื่องมือที่ช่วยให้มองเห็นว่าทรัพยากรถูกกระจายไปยังส่วนใดขององค์กรและมีความสัมพันธ์กับความสำเร็จทางธุรกิจอย่างไร การเข้าใจความสัมพันธ์เหล่านี้จะช่วยให้องค์กรสามารถตัดสินใจเกี่ยวกับการจัดสรรทรัพยากร ปรับปรุงกระบวนการ และวางแผนการเติบโตในกิจกรรมต่าง ๆ ได้อย่างมีประสิทธิภาพ

```python
activities = ["Motor Insurance", "Fire Insurance", "Property Insurance", "Accident Insurance"]
personnel_numbers = [10311, 2203, 24556, 95591]  # Number of personnel
premiums = [9384598804.05, 274180025.53, 364622255.40, 325627253.75]  # Premium amounts in Baht
policies = [4088929, 26249, 24556, 95591]  # Number of policies (used for bubble size)

# Create a scatter plot
plt.figure(figsize=(10, 6))
bubble_sizes = [p / 1000 for p in policies]  # Scale bubble sizes for better visibility
plt.scatter(personnel_numbers, premiums, s=bubble_sizes, alpha=0.6, color='green')

# Annotate the plot with activity names
for i, activity in enumerate(activities):
    plt.text(personnel_numbers[i], premiums[i], activity, fontsize=9, ha='right')

# Set axis labels and title
plt.xlabel('Number of Personnel')
plt.ylabel('Premium (Baht)')
plt.title('Comparative Analysis: Efficiency of Activities Based on Personnel and Premiums')

# Show the plot
plt.show()
```

กิจกรรมที่มีความเสี่ยงปานกลางแต่ผลตอบแทนต่ำ: การประกันภัยทรัพย์สิน (Property Insurance) อาจเป็นกิจกรรมที่ต้องการการปรับปรุงในด้านการจัดการผลตอบแทนเมื่อเทียบกับความเสี่ยงที่รับผิดชอบ

```python
# การคำนวณสัดส่วนเบี้ยประกันภัยต่อจำนวนเงินเอาประกันภัย (Premium-to-Sum Insured Ratio)

# ข้อมูล: กิจกรรม, จำนวนเงินเอาประกันภัย, และเบี้ยประกันภัย
activities = ["Motor Insurance", "Fire Insurance", "Property Insurance", "Accident Insurance"]
sum_insured = [204415210180, 130750954960, 299542192748, 974649247335]  # Sum insured amounts in Baht
premiums = [9384598804.05, 274180025.53, 364622255.40, 325627253.75]  # Premium amounts in Baht

# คำนวณสัดส่วนเบี้ยประกันภัยต่อจำนวนเงินเอาประกันภัย (Premium-to-Sum Insured Ratio)
premium_to_sum_insured_ratio = [premium / insured for premium, insured in zip(premiums, sum_insured)]

# สร้างกราฟแท่งเพื่อแสดงสัดส่วนเบี้ยประกันภัยต่อจำนวนเงินเอาประกันภัย
plt.figure(figsize=(10, 6))
plt.barh(activities, premium_to_sum_insured_ratio, color='skyblue')

# เพิ่มป้ายชื่อแกนและชื่อกราฟ
plt.xlabel('Premium-to-Sum Insured Ratio')
plt.ylabel('Activity')
plt.title('Premium-to-Sum Insured Ratio for Each Activity')

# แสดงกราฟ
plt.show()
```

การวิเคราะห์สัดส่วนเบี้ยประกันภัยต่อจำนวนเงินเอาประกันภัย (Premium-to-Sum Insured Ratio)
รายละเอียดการวิเคราะห์:
Motor Insurance (การประกันภัยรถ): มีสัดส่วนเบี้ยประกันภัยต่อจำนวนเงินเอาประกันภัยสูงที่สุด ซึ่งแสดงถึงการตั้งราคาเบี้ยประกันภัยที่ค่อนข้างสูงเมื่อเทียบกับจำนวนเงินที่เอาประกันภัย แสดงให้เห็นว่าการประกันภัยรถเป็นกิจกรรมที่บริษัทสามารถเรียกเก็บเบี้ยประกันภัยได้ในระดับที่คุ้มค่ากับความเสี่ยงที่รับ
Fire Insurance (การประกันอัคคีภัย): มีสัดส่วนเบี้ยประกันภัยต่อจำนวนเงินเอาประกันภัยที่ต่ำกว่า แต่ก็ยังคงเป็นกิจกรรมที่มีการตั้งราคาเบี้ยประกันภัยที่สอดคล้องกับความเสี่ยงที่เกี่ยวข้องกับการเกิดอัคคีภัย
Property Insurance (การประกันภัยทรัพย์สิน) และ Accident Insurance (การประกันภัยอุบัติเหตุ): มีสัดส่วนที่ค่อนข้างต่ำ แสดงว่าการประกันภัยเหล่านี้มีความเสี่ยงสูงเมื่อเทียบกับเบี้ยประกันภัยที่บริษัทสามารถเรียกเก็บได้
สรุปผลการวิเคราะห์:
กิจกรรมที่มีสัดส่วนเบี้ยประกันภัยต่อจำนวนเงินเอาประกันภัยสูง เช่น Motor Insurance อาจบ่งบอกถึงการจัดการความเสี่ยงที่ดีหรือการตั้งราคาเบี้ยประกันภัยที่เหมาะสม
กิจกรรมที่มีสัดส่วนต่ำ เช่น Property Insurance และ Accident Insurance อาจต้องมีการพิจารณาในการปรับปรุงการตั้งราคาเบี้ยประกันภัย หรือเพิ่มการจัดการความเสี่ยงเพื่อให้บริษัทได้รับผลตอบแทนที่คุ้มค่ากับความเสี่ยง

```python
# การคำนวณกำไรต่อบุคลากร (Profit per Personnel)
# ในกรณีนี้ เราจะใช้เบี้ยประกันภัยต่อจำนวนบุคลากรเป็นตัวแทนของกำไร

# ข้อมูล: กิจกรรม, จำนวนบุคลากร และเบี้ยประกันภัย
activities = ["Motor Insurance", "Fire Insurance", "Property Insurance", "Accident Insurance"]
personnel_numbers = [10311, 2203, 24556, 95591]  # Number of personnel
premiums = [9384598804.05, 274180025.53, 364622255.40, 325627253.75]  # Premium amounts in Baht

# คำนวณกำไรต่อบุคลากร (Profit per Personnel)
profit_per_personnel = [premium / personnel for premium, personnel in zip(premiums, personnel_numbers)]

# สร้างกราฟแท่งเพื่อแสดงกำไรต่อบุคลากรในแต่ละกิจกรรม
plt.figure(figsize=(10, 6))
plt.barh(activities, profit_per_personnel, color='lightgreen')

# เพิ่มป้ายชื่อแกนและชื่อกราฟ
plt.xlabel('Profit per Personnel (Baht)')
plt.ylabel('Activity')
plt.title('Profit per Personnel for Each Activity')

# แสดงกราฟ
plt.show()
```

การวิเคราะห์กำไรต่อบุคลากร (Profit per Personnel)
รายละเอียดการวิเคราะห์:
Motor Insurance (การประกันภัยรถ): มีประสิทธิภาพในการใช้บุคลากรมากที่สุด โดยมีกำไรต่อบุคลากรสูงที่สุด (ประมาณ 800,000 บาทต่อคน) แสดงว่ากิจกรรมนี้สามารถใช้บุคลากรในการสร้างรายได้ที่คุ้มค่าและมีประสิทธิภาพสูง
Fire Insurance (การประกันอัคคีภัย): มีกำไรต่อบุคลากรในระดับปานกลาง แสดงถึงการใช้ทรัพยากรบุคลากรที่ดี แต่ยังมีช่องว่างในการปรับปรุงเพื่อเพิ่มประสิทธิภาพ
Property Insurance (การประกันภัยทรัพย์สิน) และ Accident Insurance (การประกันภัยอุบัติเหตุ): มีกำไรต่อบุคลากรต่ำที่สุด แสดงถึงการใช้บุคลากรที่อาจจะไม่คุ้มค่าเท่ากับกิจกรรมอื่นๆ ซึ่งอาจจำเป็นต้องมีการปรับปรุงกระบวนการหรือเพิ่มประสิทธิภาพการทำงาน
ข้อเสนอแนะ:
Motor Insurance ควรได้รับการสนับสนุนให้ดำเนินการต่อไปในลักษณะปัจจุบัน เนื่องจากสามารถสร้างกำไรต่อบุคลากรในระดับสูง
Fire Insurance สามารถปรับปรุงการใช้ทรัพยากรบุคลากรเพื่อเพิ่มประสิทธิภาพและสร้างกำไรได้มากขึ้น
Property Insurance และ Accident Insurance ควรพิจารณาปรับปรุงกระบวนการทำงานและการจัดการทรัพยากรบุคลากร เพื่อเพิ่มกำไรต่อบุคลากรในกิจกรรมเหล่านี้
